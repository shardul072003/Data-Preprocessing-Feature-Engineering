```python
df.replace('?' , np.nan, inplace=True )
df = df.dropna()
```

```python
num_columns = df.select_dtypes(include=["float","int"]).columns
```

```python
cate_columns =df.select_dtypes(include=["object"]).columns
```

```python
num_columns
```

```python
scaler = StandardScaler()
df_std = df.copy()
```

```python
df_std[num_columns]= scaler.fit_transform(df[num_columns])
df_std[num_columns].head()
```

```python
minmax = MinMaxScaler()
df_minmax = df.copy()
```

```python
df_minmax[num_columns] = minmax.fit_transform(df[num_columns])
df_minmax[num_columns].head()
```

```python
low_cat = [c for c in cate_columns if df[c].nunique() < 5]
high_cat = [c for c in cate_columns if df[c].nunique() >= 5]
```

```python
low_cat
```

```python
high_cat
```

```python
enc = pd.get_dummies(df , columns= low_cat , drop_first=True )
```

```python
le = LabelEncoder()
```

```python
for col in high_cat:
    enc[col] = le.fit_transform(enc[col])
```

```python
### feature engineering for capital gain
enc['has_capital_gain'] = (enc['capital_gain'] > 0).astype(int)
```

```python
## feature engineering for work intensity 
enc['work_intensity'] = enc['hours_per_week'] / enc['age']
```

```python
### log transformation
enc['log_capital_gain'] = np.log1p(enc['capital_gain'])
enc[['capital_gain', 'log_capital_gain']].head()
```
